```python
import pandas as pd

# Load the CSV files
data_21_22 = pd.read_csv('21-22 data.csv')
data_22_23 = pd.read_csv('22-23 data.csv')
data_23_24 = pd.read_csv('23-24 data.csv')

# Add a column to each dataset to identify the season
data_21_22['Season'] = '21-22'
data_22_23['Season'] = '22-23'
data_23_24['Season'] = '23-24'

# Combine all data into a single DataFrame
all_data = pd.concat([data_21_22, data_22_23, data_23_24], ignore_index=True)

# Display the first few rows of the combined DataFrame
all_data.head(100)

# Remove unnecessary columns
all_data = all_data.drop(columns=['Rk', 'Awards'], errors='ignore')  # In case these columns are not present in all files

all_data.fillna(0)
```

```python
# Check for missing values in the dataset
missing_values = all_data.isnull().sum()
missing_values = missing_values[missing_values > 0]
print("Columns with missing values:\n", missing_values)
```

```python
# Fill missing values in 3P% and FT% with the median of each column
all_data['3P%'].fillna(0, inplace=True)
all_data['FT%'].fillna(0, inplace=True)

# Verify that there are no more missing values
missing_values_after = all_data.isnull().sum()
print("Missing values after handling:\n", missing_values_after[missing_values_after > 0])
```

```python


# Iterate over the data to handle '2TM', '3TM', and '4TM' entries
def replace_team_name_for_multiple_teams(player_data):
    if len(player_data) > 1:
        # Find the row where the Team is '2TM', '3TM', or '4TM'
        multiple_team_row = player_data[player_data['Team'].str.contains('TM')].iloc[0]
        
        # Get the team name from the last team row
        actual_team_name = player_data.iloc[-1]['Team']
        
        # Update the '2TM', '3TM', or '4TM' row with the actual team name
        multiple_team_row['Team'] = actual_team_name
        
        return multiple_team_row
    return player_data.iloc[0]

# Apply the logic to handle players who have multiple teams in the same season
all_data_updated = all_data.groupby(['Player', 'Season']).apply(replace_team_name_for_multiple_teams)

# Reset the index to clean up the DataFrame
all_data_updated.reset_index(drop=True, inplace=True)

# Display the cleaned and updated data
all_data_updated
```

```python
# Sort the data by Player and Season for consistency
all_data_updated.sort_values(by=['Player', 'Season'], inplace=True)

# Create a column to indicate if the player moved teams between seasons
all_data_updated['Team_Changed'] = all_data_updated.groupby('Player')['Team'].shift() != all_data_updated['Team']

# Replace NaN with False (as the first season has no previous data to compare)
all_data_updated['Team_Changed'] = all_data_updated['Team_Changed'].fillna(False)

# Display players who changed teams between any of the seasons
players_with_team_changes = all_data_updated[all_data_updated['Team_Changed']]

players_with_team_changes[['Player', 'Season', 'Team', 'Team_Changed']]
```

```python
# Perform one-hot encoding on the 'Pos' column
one_hot_encoder = pd.get_dummies(all_data_updated['Pos'], prefix='Pos')

# Concatenate the one-hot encoded columns with the original dataset
all_data_encoded = pd.concat([all_data_updated, one_hot_encoder], axis=1)

# Drop the original 'Pos' column
all_data_encoded.drop(columns=['Pos'], inplace=True)

# Verify the changes
print(all_data_encoded.head())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder


# Assuming 'all_data_updated' is your cleaned dataset with relevant columns

# Define features and targets for the linear regression models
features = ['Age', 'MP', 'G', '3P', 'AST', 'TRB', 'STL', 'BLK', 'TOV', 'PTS', '3P%', 'eFG%', 'Team_Changed', 'FT%', 'FGA']  + list(one_hot_encoder.columns) # Feel free to add more relevant features

# Create separate datasets for FG% and FT%
X_fg = all_data_encoded[features]

# FG% Model
y_fg = all_data_encoded['FG%']
X_train_fg, X_test_fg, y_train_fg, y_test_fg = train_test_split(X_fg, y_fg, test_size=0.2, random_state=42)

# Initialize and train the Linear Regression model for FG%
model_fg = LinearRegression()
model_fg.fit(X_train_fg, y_train_fg)

# Predict and evaluate the FG% model
y_pred_fg = model_fg.predict(X_test_fg)
mse_fg = mean_squared_error(y_test_fg, y_pred_fg)
r2_fg = r2_score(y_test_fg, y_pred_fg)

print(f'FG% Model Performance:')
print(f'Mean Squared Error: {mse_fg}')
print(f'R^2 Score: {r2_fg}\n')

# Display the coefficients for each model
coefficients_fg = pd.DataFrame({'Feature': features, 'Coefficient': model_fg.coef_})

print("FG% Model Coefficients:")
print(coefficients_fg)
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the hyperparameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],   # Number of trees in the forest
    'max_features': ['sqrt', 'log2', None],      # Removed 'auto', replaced with valid options
    'max_depth': [None, 10, 20, 30, 40, 50],     # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],             # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],               # Minimum samples required to be at a leaf node
    'bootstrap': [True, False]                   # Method for sampling data points (with or without replacement)
}

def tune_random_forest(X_train, y_train, X_test, y_test, category_name):
    print(f"Tuning Random Forest for {category_name}...")
    
    # Initialize RandomizedSearchCV
    random_search = RandomizedSearchCV(estimator=RandomForestRegressor(random_state=42),
                                       param_distributions=param_grid,
                                       n_iter=50,     # Number of iterations
                                       cv=3,          # 3-fold cross-validation
                                       verbose=2, 
                                       random_state=42, 
                                       n_jobs=-1)     # Use all available cores
    
    # Fit RandomizedSearchCV
    random_search.fit(X_train, y_train)
    
    # Get the best parameters and model
    best_params = random_search.best_params_
    best_rf_model = random_search.best_estimator_
    
    # Predict and evaluate the model on test data
    y_pred = best_rf_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"Best hyperparameters for {category_name}: {best_params}")
    print(f"{category_name} Model Performance after tuning:")
    print(f"Mean Squared Error: {mse}")
    print(f"R^2 Score: {r2}\n")
    
    return best_rf_model, best_params
```

```python
# Points (PTS)
best_rf_pts, best_params_pts = tune_random_forest(X_train_pts, y_train_pts, X_test_pts, y_test_pts, "PTS")

# Three-Point Shots Made (3P)
best_rf_3p, best_params_3p = tune_random_forest(X_train_3p, y_train_3p, X_test_3p, y_test_3p, "3P")

# Assists (AST)
best_rf_ast, best_params_ast = tune_random_forest(X_train_ast, y_train_ast, X_test_ast, y_test_ast, "AST")

# Rebounds (REB)
best_rf_reb, best_params_reb = tune_random_forest(X_train_reb, y_train_reb, X_test_reb, y_test_reb, "REB")

# Field Goal Percentage (FG%)
best_rf_fg, best_params_fg = tune_random_forest(X_train_fg, y_train_fg, X_test_fg, y_test_fg, "FG%")

# Free Throw Percentage (FT%)
best_rf_ft, best_params_ft = tune_random_forest(X_train_ft, y_train_ft, X_test_ft, y_test_ft, "FT%")

# Blocks (BLK)
best_rf_blk, best_params_blk = tune_random_forest(X_train_blk, y_train_blk, X_test_blk, y_test_blk, "BLK")

# Steals (STL)
best_rf_stl, best_params_stl = tune_random_forest(X_train_stl, y_train_stl, X_test_stl, y_test_stl, "STL")

# Turnovers (TO)
best_rf_to, best_params_to = tune_random_forest(X_train_to, y_train_to, X_test_to, y_test_to, "TO")
```

```python
# Assuming predictions_df contains the projected stats and games played data
# Calculate z-scores for each category
categories = ['Projected_PTS', 'Projected_3P', 'Projected_AST', 'Projected_REB', 'Projected_FG%', 
              'Projected_FT%', 'Projected_BLK', 'Projected_STL', 'Projected_TO']

# Calculate z-scores for each category
for category in categories:
    predictions_df[f'{category}_z'] = (predictions_df[category] - predictions_df[category].mean()) / predictions_df[category].std()

# Multiply the turnover (TO) z-scores by -1 since lower turnovers are better in fantasy basketball
predictions_df['Projected_TO_z'] *= -1

# Calculate the total z-score by summing all category z-scores
predictions_df['Total_z'] = (
    predictions_df['Projected_PTS_z'] +
    predictions_df['Projected_3P_z'] +
    predictions_df['Projected_AST_z'] +
    predictions_df['Projected_REB_z'] +
    predictions_df['Projected_FG%_z'] +
    predictions_df['Projected_FT%_z'] +
    predictions_df['Projected_BLK_z'] +
    predictions_df['Projected_STL_z'] +
    predictions_df['Projected_TO_z']
)

# Define a minimum threshold for expected games played, e.g., 65 games for second-year players
min_games_threshold = 65

# Identify players who are second-year or have limited data (G < 40) and adjust their projected games
# Assuming that `Age` represents their age for the upcoming season
# Adjust for players who are likely to have more games played if they are young and had low games last season
predictions_df['Adjusted_G'] = predictions_df.apply(
    lambda row: max(row['G'], min_games_threshold) if (row['G'] < min_games_threshold and row['Age'] <= 22) else row['G'],
    axis=1
)

# Recalculate Fantasy Value using Adjusted_G
predictions_df['Fantasy_Value'] = predictions_df['Total_z'] * predictions_df['Adjusted_G']

# Rank players based on the adjusted fantasy value
predictions_df['Draft_Rank'] = predictions_df['Fantasy_Value'].rank(ascending=False)

# Sort by the draft rank to get the final ranking
draft_ranking_df = predictions_df.sort_values(by='Draft_Rank')

# Round all relevant numerical columns to 2 decimal places in the draft_ranking_df
draft_ranking_df = draft_ranking_df.round(2)

```
